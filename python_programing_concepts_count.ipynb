```python
# prompt: can you find out basic stats information on my dataframe df, calculate measurements such as mean, median, mode, variance, std, range, skewness, kurtosis

import pandas as pd
from scipy.stats import skew, kurtosis

# Assuming 'df' is your DataFrame and 'salary' is the column you want to analyze
# If your column name is different replace 'salary' with your column name

# Mean
mean_salary = df['salary'].mean()
print("Mean:", mean_salary)

# Median
median_salary = df['salary'].median()
print("Median:", median_salary)

# Mode
mode_salary = df['salary'].mode()  #access the first mode if multipal exists
print("Mode:", mode_salary)

# Variance
variance_salary = df['salary'].var()
print("Variance:", variance_salary)

# Standard Deviation
std_salary = df['salary'].std()
print("Standard Deviation:", std_salary)

# Range
range_salary = df['salary'].max() - df['salary'].min() #range=max-min
print("Range:", range_salary)

# Skewness
skewness_salary = skew(df['salary'])
print("Skewness:", skewness_salary)

# Kurtosis
kurtosis_salary = kurtosis(df['salary'])
print("Kurtosis:", kurtosis_salary)
```

```python
# prompt: can you create distplot for all the columns in dataframe df1

import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df1 is your DataFrame
for col in df1.columns:
    if df1[col].dtype in ['int64', 'float64']:  # Check for numeric columns
        plt.figure(figsize=(8, 6))  # Adjust figure size as needed
        sns.distplot(df1[col])
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.show()
```
